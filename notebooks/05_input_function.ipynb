#### create a input function to predict the price

```python
import joblib
import pandas as pd
```

```python
model_package = joblib.load("../models/car_price_model_v2.pkl")
rf=model_package['model']
features=model_package["feature"]
```

```python
def predict_car_price(age, km, weight, hp):
    # Create dataframe in correct feature order
    input_data = pd.DataFrame([[age, km, weight, hp]], columns=features)
    
    # Predict
    predicted_price = rf.predict(input_data)[0]
    
    return round(predicted_price, 2)
```

```python
price = predict_car_price(
    age=5,        # years old
    km=60000,     # kilometers driven
    weight=1200,  # kg
    hp=110        # horsepower
)

print("Estimated Car Price:", price)
```
